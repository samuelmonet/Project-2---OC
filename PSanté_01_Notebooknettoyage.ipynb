# Traitement des doublons

```python
#Recherche des doublons

data3=data2.copy()
#Je cherche les produit ayant meme code barre
dupl=data3[data3.duplicated(subset='code',keep=False)].sort_values(by='code')
dupl
```

```python

#Je complète les éventuelles cellules vides par les valeurs du doublon
for i in dupl.code.unique():
    na=data3[data3['code']==i].isnull()
    L=na.index.tolist()
    for k in na.columns:
        for j in range(2):
            if na.loc[L[j],k]:
                if not na.loc[L[(j+1)%2],k]:
                    data3.loc[L[j],k]=data3.loc[L[(j+1)%2], k]
#Je supprime les doublons sur les codes barres:
data3.drop_duplicates(subset='code',inplace=True)
print(data2.shape,data3.shape)
                    
```

# Traitement des outliers

```python


import matplotlib
gs = matplotlib.gridspec.GridSpec(1, 2,width_ratios = [1, 8])

#On va représenter ces différentes variables dans un diagramme en boite:
fig=plt.figure(figsize=(15,10))

#On en fait une spéciale énergie car les valeurs y sont particulièrement plus grandes:

plt.subplot(gs[0,0])
e=plt.boxplot([data3['energy_100g'].dropna(how='any')],showfliers=True,positions=[16],widths=3)


plt.gca().xaxis.set_ticklabels(['energy'],size=12,rotation=90)
for element in e['medians']:
    element.set_color('blue')
    element.set_linewidth(3)
for element in e['means']:
    element.set_color('yellow')
    element.set_linewidth(3)
for element in e['caps']:
    element.set_color('red')
    element.set_linewidth(3)

plt.xlim(12,22)
plt.ylim(0,3000)


plt.subplot(gs[0,1])

a=plt.boxplot([data3[g].dropna(how='any') for g in data3.columns.tolist()[4:]],\
              positions= [10*i+35 for i in range(len(data3.columns.tolist()[4:]))],showfliers= True,widths= 3)
              
plt.gca().xaxis.set_ticklabels([i for i in name(data3.columns.tolist()[4:])],size=12,rotation=90)
for element in a['medians']:
    element.set_color('blue')
    element.set_linewidth(2)
for element in a['caps']:
    element.set_color('red')
    element.set_linewidth(2)
for element in a['means']:
    element.set_color('orange')
    element.set_linewidth(2)
    
plt.xlim(30,110)
plt.ylim(0,100)





plt.suptitle('Diagrammes en boite avant nettoyage',size=22)
plt.show()
```
